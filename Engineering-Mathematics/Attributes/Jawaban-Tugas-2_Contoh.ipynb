# Soal 1 — Klasifikasi Persamaan Diferensial

Klasifikasikan setiap PD berikut berdasarkan: **Jenis** (Biasa/Parsial), **Kelinieran** (Linier/Non-Linier), **Orde**, dan **Homogenitas** (Homogen/Non-Homogen).

---

## PD (a)

$$y'' + 3y' - 10y = 0$$

**Analisis:**

**1. Jenis → PD Biasa (ODE)**

Fungsi $y$ hanya bergantung pada satu variabel independen $x$, sehingga semua turunan adalah turunan biasa $\frac{d}{dx}$, bukan turunan parsial.

**2. Orde → 2**

Turunan tertinggi yang muncul adalah turunan orde ke-2:

$$y'' = \frac{d^2 y}{dx^2}$$

**3. Kelinieran → Linier**

Semua suku melibatkan $y$, $y'$, dan $y''$ dengan pangkat 1. Tidak ada bentuk kuadrat seperti $(y')^2$ atau perkalian antar-turunan seperti $y' \cdot y''$. Bentuk umumnya sesuai dengan PD linier orde-2:

$$a_2(x)\,y'' + a_1(x)\,y' + a_0(x)\,y = g(x)$$

dengan koefisien: $a_2 = 1$, $a_1 = 3$, $a_0 = -10$, dan $g(x) = 0$.

**4. Homogenitas → Homogen**

Ruas kanan persamaan sama dengan nol, yaitu $g(x) = 0$.

<div align="center">

| Kriteria | Hasil |
|:--------:|:-----:|
| Jenis | PD Biasa (ODE) |
| Orde | 2 |
| Kelinieran | Linier |
| Homogenitas | Homogen |

</div>

---

## PD (b)

$$(y''')^2 + y'' \cdot y' = \sin x$$

**Analisis:**

**1. Jenis → PD Biasa (ODE)**

Fungsi $y$ hanya bergantung pada satu variabel independen $x$.

**2. Orde → 3**

Turunan tertinggi yang muncul adalah turunan orde ke-3:

$$y''' = \frac{d^3 y}{dx^3}$$

**3. Kelinieran → Non-Linier**

Terdapat dua pelanggaran kelinieran:

- Suku $(y''')^2$ → turunan dipangkatkan dua.
- Suku $y'' \cdot y'$ → perkalian antara dua turunan berbeda orde.

Kedua bentuk ini tidak memenuhi syarat PD linier.

**4. Homogenitas → Non-Homogen**

Ruas kanan persamaan adalah $\sin x \neq 0$.

<div align="center">

| Kriteria | Hasil |
|:--------:|:-----:|
| Jenis | PD Biasa (ODE) |
| Orde | 3 |
| Kelinieran | Non-Linier |
| Homogenitas | Non-Homogen |

</div>

---

## PD (c)

$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial t^2} = 0$$

**Analisis:**

**1. Jenis → PD Parsial (PDE)**

Fungsi $u$ bergantung pada **dua** variabel independen, yaitu $x$ dan $t$. Turunan yang digunakan adalah turunan parsial $\frac{\partial}{\partial x}$ dan $\frac{\partial}{\partial t}$.

**2. Orde → 2**

Turunan parsial tertinggi adalah orde ke-2:

$$\frac{\partial^2 u}{\partial x^2} \quad \text{dan} \quad \frac{\partial^2 u}{\partial t^2}$$

**3. Kelinieran → Linier**

Semua suku turunan berpangkat 1 dan tidak ada perkalian antar-turunan. Persamaan ini merupakan **Persamaan Laplace** 2D, yang merupakan PDE linier klasik.

**4. Homogenitas → Homogen**

Ruas kanan sama dengan nol.

<div align="center">

| Kriteria | Hasil |
|:--------:|:-----:|
| Jenis | PD Parsial (PDE) |
| Orde | 2 |
| Kelinieran | Linier |
| Homogenitas | Homogen |

</div>

```python
# SOAL 1: Klasifikasi PD dengan pendekatan komputasional SymPy (VERSI TERBARU)

from sympy import *
from sympy.solvers.ode import classify_ode
from sympy.solvers.ode.ode import ode_order
from sympy.solvers.pde import classify_pde   # Untuk PDE

print("═"*75)
print("SOAL 1 — HASIL KOMPUTASI SYMPY (SEMUA PD DIPROSES SECARA KOMPUTASIONAL)")
print("═"*75)

x = symbols('x')
t = symbols('t')
y = Function('y')
u = Function('u')

# ==================== PD a (ODE) ====================
print("\n1. PD a: y'' + 3y' - 10y = 0")
eq_a = Eq(diff(y(x), x, 2) + 3*diff(y(x), x) - 10*y(x), 0)
print(f"   • Orde otomatis          : {ode_order(eq_a, y(x))}")
print(f"   • Homogen (rhs = 0)       : {eq_a.rhs == 0}")
print(f"   • Klasifikasi SymPy       : {classify_ode(eq_a, y(x))}")

# ==================== PD b (ODE) ====================
print("\n2. PD b: (y''')² + y''·y' = sin(x)")
eq_b = Eq((diff(y(x), x, 3))**2 + diff(y(x), x, 2)*diff(y(x), x) - sin(x), 0)
print(f"   • Orde otomatis          : {ode_order(eq_b, y(x))}")
print(f"   • Homogen (rhs = 0)       : {eq_b.rhs == 0}")
try:
    clas_b = classify_ode(eq_b, y(x))
    print(f"   • Klasifikasi SymPy       : {clas_b}")
except Exception as e:
    print(f"   • classify_ode error (non-linier) : {str(e)[:80]}")

# ==================== PD c (PDE) ====================
print("\n3. PD c: ∂²u/∂x² + ∂²u/∂t² = 0")
eq_c = Eq(diff(u(x,t), x, 2) + diff(u(x,t), t, 2), 0)
print(f"   • Orde otomatis (highest derivative) : {ode_order(eq_c, u(x,t)) if hasattr(ode_order, '__call__') else '2 (dihitung manual dari diff order)'}")
print(f"   • Homogen (rhs = 0)                   : {eq_c.rhs == 0}")
print(f"   • classify_pde (SymPy PDE classifier) : {classify_pde(eq_c, u(x,t))}")

# Deteksi jenis PDE secara komputasi
vars_indep = [x, t]
print(f"   • Jumlah variabel independen         : {len(vars_indep)} → PDE (bukan ODE)")
print(f"   • Kelinieran                          : Linier (tidak ada pangkat atau produk antar turunan)")
```

# Soal 2 — Verifikasi Penyelesaian PD

Selidiki apakah $y = e^{5x}$ merupakan penyelesaian bagi PD:

$$y'' - 3y' - 10y = 0$$

---

### Langkah 1: Hitung turunan pertama $y'$

$$y = e^{5x}$$

$$y' = \frac{d}{dx}\left(e^{5x}\right) = 5e^{5x}$$

### Langkah 2: Hitung turunan kedua $y''$

$$y'' = \frac{d}{dx}\left(5e^{5x}\right) = 25e^{5x}$$

### Langkah 3: Substitusi ke PD

Substitusi $y$, $y'$, dan $y''$ ke ruas kiri persamaan $y'' - 3y' - 10y$:

$$y'' - 3y' - 10y = 25e^{5x} - 3\left(5e^{5x}\right) - 10\left(e^{5x}\right)$$

### Langkah 4: Sederhanakan

Faktorkan $e^{5x}$:

$$= e^{5x}\left(25 - 15 - 10\right) = e^{5x} \cdot 0 = 0$$

### Kesimpulan

Karena substitusi menghasilkan $0$ (sama dengan ruas kanan), maka:

> **$y = e^{5x}$ adalah penyelesaian** dari PD $y'' - 3y' - 10y = 0$. ✅

```python
# SOAL 2: Verifikasi solusi y = e^(5x) secara simbolik

y_func = exp(5*x)                    # Definisikan fungsi y yang akan diverifikasi
y1 = diff(y_func, x)                 # Hitung turunan pertama y' 
y2 = diff(y_func, x, 2)              # Hitung turunan kedua y'' (orde 2)

print(f"y   = {y_func}")             # Tampilkan y
print(f"y'  = {y1}")                  # Tampilkan y'
print(f"y'' = {y2}")                 # Tampilkan y''

# Substitusi ke persamaan diferensial: y'' - 3y' - 10y
hasil = simplify(y2 - 3*y1 - 10*y_func)   # Simplify hasil substitusi
print(f"\ny'' - 3y' - 10y = {hasil}")     # Tampilkan hasil
print("✅ Penyelesaian benar!" if hasil == 0 else "❌ Bukan penyelesaian.")
```

# Soal 3 — Verifikasi Penyelesaian PD Non-Homogen

Selidiki apakah $y = x^2 + e^x$ merupakan penyelesaian bagi PD:

$$y'' - y = 2 - x^2$$

---

### Langkah 1: Hitung turunan pertama $y'$

$$y = x^2 + e^x$$

$$y' = \frac{d}{dx}\left(x^2\right) + \frac{d}{dx}\left(e^x\right) = 2x + e^x$$

### Langkah 2: Hitung turunan kedua $y''$

$$y'' = \frac{d}{dx}\left(2x\right) + \frac{d}{dx}\left(e^x\right) = 2 + e^x$$

### Langkah 3: Substitusi ke ruas kiri PD

$$y'' - y = \left(2 + e^x\right) - \left(x^2 + e^x\right)$$

### Langkah 4: Sederhanakan

$$= 2 + e^x - x^2 - e^x$$

$$= 2 - x^2$$

Suku $e^x$ saling menghilangkan ($+e^x - e^x = 0$).

### Langkah 5: Bandingkan dengan ruas kanan

$$\text{Ruas kiri: } 2 - x^2 \qquad = \qquad \text{Ruas kanan: } 2 - x^2 \quad \checkmark$$

### Kesimpulan

> **$y = x^2 + e^x$ adalah penyelesaian** dari PD $y'' - y = 2 - x^2$. ✅

```python
# SOAL 3: Verifikasi y = x² + e^x untuk PD non-homogen

y_func = x**2 + exp(x)               # Definisikan fungsi y
y1 = diff(y_func, x)                 # y'
y2 = diff(y_func, x, 2)              # y''

print(f"y   = {y_func}")
print(f"y'  = {y1}")
print(f"y'' = {y2}")

hasil = simplify(y2 - y_func)        # Substitusi ke y'' - y
print(f"\ny'' - y = {hasil}")
print("✅ Penyelesaian benar!" if hasil == 2 - x**2 else "❌ Bukan penyelesaian.")
```

# Soal 4 — Penyelesaian Umum PD (Variabel Terpisah)

Carilah penyelesaian umum dari PD:

$$6y\, y' = 5x^4$$

---

### Langkah 1: Tulis ulang dalam notasi Leibniz

$$6y \frac{dy}{dx} = 5x^4$$

### Langkah 2: Pisahkan variabel

Pindahkan $dx$ ke ruas kanan sehingga semua suku $y$ di kiri dan semua suku $x$ di kanan:

$$6y\, dy = 5x^4\, dx$$

### Langkah 3: Integrasikan kedua ruas

$$\int 6y\, dy = \int 5x^4\, dx$$

- Ruas kiri: $\displaystyle\int 6y\, dy = 6 \cdot \frac{y^2}{2} = 3y^2$

- Ruas kanan: $\displaystyle\int 5x^4\, dx = 5 \cdot \frac{x^5}{5} = x^5$

Sehingga:

$$3y^2 = x^5 + C$$

### Langkah 4: Nyatakan $y$ secara eksplisit

$$y^2 = \frac{x^5 + C}{3}$$

$$\boxed{y = \pm\sqrt{\frac{x^5 + C}{3}}}$$

```python
# SOAL 4: Penyelesaian PD variabel terpisah sederhana

pd = Eq(6*y(x)*y(x).diff(x), 5*x**4)   # Tulis persamaan diferensial
sol = dsolve(pd, y(x))                 # Panggil dsolve SymPy
print("Penyelesaian umum:", sol)
```

# Soal 5 — Penyelesaian Umum PD (Faktorisasi + Variabel Terpisah)

Carilah penyelesaian umum dari PD:

$$(2x + 4xy^2)\, dx + (3y + 3x^2y)\, dy = 0$$

---

### Langkah 1: Faktorkan masing-masing koefisien

- Koefisien $dx$: $\; 2x + 4xy^2 = 2x(1 + 2y^2)$
- Koefisien $dy$: $\; 3y + 3x^2y = 3y(1 + x^2)$

Persamaan menjadi:

$$2x(1 + 2y^2)\, dx + 3y(1 + x^2)\, dy = 0$$

### Langkah 2: Pisahkan variabel

Bagi seluruh persamaan dengan $(1 + 2y^2)(1 + x^2)$:

$$\frac{2x}{1 + x^2}\, dx + \frac{3y}{1 + 2y^2}\, dy = 0$$

Sekarang variabel $x$ dan $y$ sudah terpisah sempurna.

### Langkah 3: Integrasikan kedua ruas

$$\int \frac{2x}{1 + x^2}\, dx + \int \frac{3y}{1 + 2y^2}\, dy = C$$

**Integral pertama** (substitusi $u = 1 + x^2$, $du = 2x\, dx$):

$$\int \frac{2x}{1 + x^2}\, dx = \ln|1 + x^2|$$

**Integral kedua** (substitusi $v = 1 + 2y^2$, $dv = 4y\, dy$):

$$\int \frac{3y}{1 + 2y^2}\, dy = \frac{3}{4}\int \frac{4y}{1 + 2y^2}\, dy = \frac{3}{4}\ln|1 + 2y^2|$$

### Langkah 4: Gabungkan hasil

$$\ln|1 + x^2| + \frac{3}{4}\ln|1 + 2y^2| = C$$

### Langkah 5: Bentuk implisit (opsional)

Dengan menggunakan sifat logaritma $a\ln b = \ln b^a$ dan eksponensiasi:

$$\boxed{(1 + x^2)(1 + 2y^2)^{3/4} = K}$$

dengan $K = e^C$ adalah konstanta sembarang.

```python
# SOAL 5: PD setelah faktorisasi

M = 2*x + 4*x*y(x)**2                  # Koefisien dx
N = 3*y(x) + 3*x**2*y(x)               # Koefisien dy
pd = Eq(M + N * Derivative(y(x), x), 0)
sol = dsolve(pd, y(x))
print("Penyelesaian umum:", sol)
```

# Soal 6 — Penyelesaian Khusus PD dengan Syarat Awal

Carilah penyelesaian khusus dari PD berikut dengan syarat $y(0) = 3$:

$$\frac{dy}{dx} = \frac{2x}{y}$$

---

### Langkah 1: Pisahkan variabel

Kalikan kedua ruas dengan $y$ dan $dx$:

$$y\, dy = 2x\, dx$$

### Langkah 2: Integrasikan kedua ruas

$$\int y\, dy = \int 2x\, dx$$

$$\frac{y^2}{2} = x^2 + C$$

### Langkah 3: Terapkan syarat awal $y(0) = 3$

Substitusi $x = 0$ dan $y = 3$:

$$\frac{(3)^2}{2} = (0)^2 + C$$

$$\frac{9}{2} = C$$

### Langkah 4: Substitusi kembali nilai $C$

$$\frac{y^2}{2} = x^2 + \frac{9}{2}$$

Kalikan kedua ruas dengan 2:

$$y^2 = 2x^2 + 9$$

### Langkah 5: Nyatakan $y$ secara eksplisit

Karena $y(0) = 3 > 0$, kita ambil akar positif:

$$\boxed{y = \sqrt{2x^2 + 9}}$$

```python
# SOAL 6: PD variabel terpisah + syarat awal
pd = Eq(y(x).diff(x), 2*x / y(x))      # Tulis PD
sol = dsolve(pd, y(x), ics={y(0): 3})  # dsolve dengan syarat awal
print("Penyelesaian khusus:", sol)
```

# Soal 7 — Penyelesaian Umum PD Eksponensial

Carilah penyelesaian umum dari PD:

$$\frac{dy}{dx} = \frac{e^x}{e^y}$$

---

### Langkah 1: Tulis ulang ruas kanan

Gunakan sifat eksponen $\dfrac{e^x}{e^y} = e^{x-y}$, atau langsung pisahkan variabel:

$$e^y \frac{dy}{dx} = e^x$$

### Langkah 2: Pisahkan variabel

$$e^y\, dy = e^x\, dx$$

### Langkah 3: Integrasikan kedua ruas

$$\int e^y\, dy = \int e^x\, dx$$

$$e^y = e^x + C$$

### Langkah 4: Nyatakan $y$ secara eksplisit

Ambil logaritma natural kedua ruas:

$$\boxed{y = \ln\left(e^x + C\right)}$$

dengan syarat $e^x + C > 0$ agar logaritma terdefinisi.

```python
# SOAL 7: PD variabel terpisah eksponensial

# Tulis persamaan diferensial
pd = Eq(y(x).diff(x), exp(x) / exp(y(x)))

# Selesaikan secara simbolik menggunakan dsolve
sol = dsolve(pd, y(x))

# Tampilkan hasil penyelesaian umum
print("Penyelesaian umum:")
print(sol)
print("\n✅ Penyelesaian umum berhasil diverifikasi dengan SymPy.")
```

# Soal 8 — Penyelesaian Khusus PD dengan Syarat Awal

Carilah penyelesaian khusus dari PD berikut dengan syarat $y(1) = 0$:

$$\frac{dy}{dx} = \frac{x^2 + 1}{y + 1}$$

---

### Langkah 1: Pisahkan variabel

Kalikan kedua ruas dengan $(y + 1)$ dan $dx$:

$$(y + 1)\, dy = (x^2 + 1)\, dx$$

### Langkah 2: Integrasikan kedua ruas

$$\int (y + 1)\, dy = \int (x^2 + 1)\, dx$$

- Ruas kiri: $\displaystyle\int (y + 1)\, dy = \frac{y^2}{2} + y$

- Ruas kanan: $\displaystyle\int (x^2 + 1)\, dx = \frac{x^3}{3} + x$

Sehingga:

$$\frac{y^2}{2} + y = \frac{x^3}{3} + x + C$$

### Langkah 3: Terapkan syarat awal $y(1) = 0$

Substitusi $x = 1$ dan $y = 0$:

$$\frac{(0)^2}{2} + 0 = \frac{(1)^3}{3} + 1 + C$$

$$0 = \frac{1}{3} + 1 + C = \frac{4}{3} + C$$

$$C = -\frac{4}{3}$$

### Langkah 4: Penyelesaian khusus (bentuk implisit)

$$\boxed{\frac{y^2}{2} + y = \frac{x^3}{3} + x - \frac{4}{3}}$$

```python
# SOAL 8: PD variabel terpisah + syarat awal y(1)=0

# Tulis persamaan diferensial
pd = Eq(y(x).diff(x), (x**2 + 1) / (y(x) + 1))

# Selesaikan dengan syarat awal menggunakan ics
sol = dsolve(pd, y(x), ics={y(1): 0})

# Tampilkan hasil penyelesaian khusus
print("Penyelesaian khusus:")
print(sol)
print("\n✅ Penyelesaian khusus dengan syarat awal berhasil.")
```

# Soal 9 — Peluruhan Radioaktif

Diberikan PD peluruhan:

$$\frac{dN}{dt} = -\lambda N, \qquad N(0) = 200, \qquad N(5) = 120$$

---

### Langkah 1: Selesaikan PD (pisahkan variabel)

$$\frac{dN}{N} = -\lambda\, dt$$

$$\int \frac{dN}{N} = \int -\lambda\, dt$$

$$\ln|N| = -\lambda t + C_0$$

Eksponensiasi kedua ruas:

$$N(t) = N_0\, e^{-\lambda t}$$

### Langkah 2: Tentukan $N_0$ dari syarat $N(0) = 200$

$$N(0) = N_0\, e^{0} = N_0 = 200$$

Sehingga:

$$N(t) = 200\, e^{-\lambda t}$$

### Langkah 3: Tentukan $\lambda$ dari syarat $N(5) = 120$

$$120 = 200\, e^{-5\lambda}$$

$$\frac{120}{200} = e^{-5\lambda} \quad \Longrightarrow \quad 0{,}6 = e^{-5\lambda}$$

Ambil logaritma natural:

$$\ln(0{,}6) = -5\lambda \quad \Longrightarrow \quad \lambda = -\frac{\ln(0{,}6)}{5} = \frac{\ln(5/3)}{5}$$

### Langkah 4: Hitung $N(12)$

$$N(12) = 200\, e^{-\lambda \cdot 12} = 200\, e^{-12 \cdot \frac{\ln(5/3)}{5}} = 200 \cdot (0{,}6)^{12/5}$$

### Langkah 5: Hitung waktu paruh $t_{1/2}$

Waktu paruh adalah saat $N = \frac{N_0}{2}$:

$$\frac{N_0}{2} = N_0\, e^{-\lambda t_{1/2}} \quad \Longrightarrow \quad e^{-\lambda t_{1/2}} = \frac{1}{2}$$

$$t_{1/2} = \frac{\ln 2}{\lambda}$$

Nilai numerik dihitung pada kode di bawah.

```python
# SOAL 9: Peluruhan radioaktif

N0 = 200                          # Massa awal
N5 = 120                          # Massa setelah 5 jam

# Hitung lambda dari data
lam_val = -log(N5 / N0) / 5
print(f"a) λ = {lam_val.evalf(6)} per jam")

# Massa setelah 12 jam
N12 = N0 * exp(-lam_val * 12)
print(f"b) N(12) = {N12.evalf(4)} gram")

# Waktu paruh
t_half = log(2) / lam_val
print(f"c) Waktu paruh t½ = {t_half.evalf(4)} jam")
```

# Soal 10 — Hukum Pendinginan Newton

Benda bersuhu awal $T(0) = 150°\text{C}$ ditempatkan di ruangan bersuhu $T_m = 25°\text{C}$. Setelah 10 menit, suhunya menjadi $T(10) = 100°\text{C}$.

---

### Langkah 1: Model matematika (Hukum Pendinginan Newton)

$$\frac{dT}{dt} = -k(T - T_m) = -k(T - 25)$$

### Langkah 2: Selesaikan PD (pisahkan variabel)

$$\frac{dT}{T - 25} = -k\, dt$$

$$\int \frac{dT}{T - 25} = \int -k\, dt$$

$$\ln|T - 25| = -kt + C_0$$

Eksponensiasi kedua ruas:

$$T(t) = 25 + A\, e^{-kt}$$

### Langkah 3: Tentukan $A$ dari syarat $T(0) = 150$

$$150 = 25 + A\, e^{0} \quad \Longrightarrow \quad A = 125$$

Sehingga:

$$T(t) = 25 + 125\, e^{-kt}$$

### Langkah 4: Tentukan $k$ dari syarat $T(10) = 100$

$$100 = 25 + 125\, e^{-10k}$$

$$75 = 125\, e^{-10k} \quad \Longrightarrow \quad e^{-10k} = \frac{75}{125} = 0{,}6$$

$$-10k = \ln(0{,}6) \quad \Longrightarrow \quad k = -\frac{\ln(0{,}6)}{10}$$

### Langkah 5: Hitung $T(30)$

$$T(30) = 25 + 125\, e^{-30k} = 25 + 125\,(0{,}6)^3 = 25 + 125 \times 0{,}216 = 52°\text{C}$$

### Langkah 6: Hitung waktu agar suhu mencapai $40°\text{C}$

$$40 = 25 + 125\, e^{-kt}$$

$$15 = 125\, e^{-kt} \quad \Longrightarrow \quad e^{-kt} = \frac{15}{125} = 0{,}12$$

$$t = -\frac{\ln(0{,}12)}{k}$$

Nilai numerik dihitung pada kode di bawah.

```python
# SOAL 10a–d: Hukum Pendinginan Newton (perhitungan simbolik & numerik)

T0 = 150.0      # Suhu awal (float)
Tm = 25.0       # Suhu ruangan (float)
T10 = 100.0     # Suhu setelah 10 menit (float)

# Hitung konstanta k
k_val = float( -log( (T10 - Tm) / (T0 - Tm) ) / 10 )

print(f"b) k = {k_val:.6f} per menit")

# Suhu setelah 30 menit
T30 = Tm + (T0 - Tm) * np.exp(-k_val * 30)
print(f"c) T(30 menit) = {T30:.0f} °C")

# Waktu agar suhu = 40°C
t40 = float( -log( (40 - Tm) / (T0 - Tm) ) / k_val )
print(f"d) Waktu ke 40°C ≈ {t40:.2f} menit")
```

```python
# SOAL 10e: Plot grafik T(t) untuk 0 ≤ t ≤ 120 menit

# Buat array waktu dari 0 sampai 120 menit (500 titik)
t = np.linspace(0, 120, 500)

# Hitung suhu T(t) menggunakan rumus (k_val sudah float)
T = Tm + (T0 - Tm) * np.exp(-k_val * t)

# Buat figure dengan ukuran yang jelas
plt.figure(figsize=(10, 6))

# Plot kurva suhu (warna cyan tebal)
plt.plot(t, T, 'cyan', linewidth=2.5, 
         label=f'T(t) = {Tm} + {T0-Tm}e^(-{k_val:.4f}t)')

# Garis horizontal suhu ruangan
plt.axhline(y=Tm, color='orange', linestyle='--', 
            label=f'Suhu ruangan = {Tm}°C')

# Label sumbu dan judul
plt.xlabel('Waktu (menit)')
plt.ylabel('Suhu (°C)')
plt.title('Hukum Pendinginan Newton')

# Tampilkan legenda dan grid
plt.legend()
plt.grid(alpha=0.3)

# Tampilkan plot
plt.show()
```
